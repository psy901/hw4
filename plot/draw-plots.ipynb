```python
# Read the data
import pandas as pd
reward_easy_df = pd.read_csv('./easy-reward.csv')
reward_hard_df = pd.read_csv('./hard-reward.csv')
```

```python
# Parse the data
reward_easy_dict = {}

for row in reward_easy_df.iterrows():
    name = row[1][0]
    rewards = row[1][1:]
    reward_easy_dict[name] = rewards
    
reward_hard_dict = {}
for row in reward_hard_df.iterrows():
    name = row[1][0]
    rewards = row[1][1:]
    reward_hard_dict[name] = rewards
```

```python
# Draw the plot
import matplotlib.pyplot as plt

colors = ['red', 'blue', 'green']
namedict = {}
for name in reward_easy_dict:
    if 'V' in name:
        namedict[name] = 'vi'
    elif 'P' in name:
        namedict[name] = 'pi'
    elif 'Q' in name:
        namedict[name] = 'q'
        
print(namedict)

for i, name in enumerate(reward_easy_dict):
    plt.clf()
    plt.plot(list(reward_easy_dict[name]), c=colors[i])
#     plt.plot([0, 100], [0, 0], c='black', linestyle='--')
    plt.xlabel('iterations', fontsize=20)
    plt.ylabel('rewards', fontsize=20)
    plt.title(name, fontsize=20)
    plt.savefig('../fig/easy-{}-reward.pdf'.format(namedict[name]))
    

for i, name in enumerate(reward_hard_dict):
    plt.clf()
    plt.plot(list(reward_hard_dict[name]), c=colors[i])
#     plt.plot([0, 100], [0, 0], c='black', linestyle='--')
    plt.xlabel('iterations', fontsize=20)
    plt.ylabel('rewards', fontsize=20)
    plt.title(name, fontsize=20)
    plt.savefig('../fig/hard-{}-reward.pdf'.format(namedict[name]))
    
```

```python
# Read the data
import pandas as pd
time_easy_df = pd.read_csv('./easy-time.csv')
time_hard_df = pd.read_csv('./hard-time.csv')
time_hard_notrap_df = pd.read_csv('./hard-notrap-time.csv')
```

```python
# Parse the data
time_easy_dict = {}

for row in time_easy_df.iterrows():
    name = row[1][0]
    times = row[1][1:]
    time_easy_dict[name] = times
    
time_hard_dict = {}
for row in time_hard_df.iterrows():
    name = row[1][0]
    times = row[1][1:]
    time_hard_dict[name] = times
    
time_hard_notrap_dict = {}
for row in time_hard_notrap_df.iterrows():
    name = row[1][0]
    times = row[1][1:]
    time_hard_notrap_dict[name] = times
```

```python
# Draw the plot
import matplotlib.pyplot as plt

colors = ['red', 'blue', 'green']
namedict = {}
for name in time_easy_dict:
    if 'V' in name:
        namedict[name] = 'vi'
    elif 'P' in name:
        namedict[name] = 'pi'
    elif 'Q' in name:
        namedict[name] = 'q'
        
print(namedict)

for i, name in enumerate(time_easy_dict):
    plt.clf()
    plt.plot(list(time_easy_dict[name]), c=colors[i])
    plt.xlabel('iterations', fontsize=20)
    plt.ylabel('execution time (milliseconds)', fontsize=20)
    plt.title(name, fontsize=20)
    plt.savefig('../fig/easy-{}-time.pdf'.format(namedict[name]))
    

for i, name in enumerate(time_hard_dict):
    plt.clf()
    plt.plot(list(time_hard_dict[name]), c=colors[i])
    plt.xlabel('iterations', fontsize=20)
    plt.ylabel('execution time (milliseconds)', fontsize=20)
    plt.title(name, fontsize=20)
    plt.savefig('../fig/hard-{}-time.pdf'.format(namedict[name]))

    
plt.clf()
plt.xlabel('iterations', fontsize=20)
plt.ylabel('execution time (milliseconds)', fontsize=20)
plt.title('with trap', fontsize=20)
for i, name in enumerate(time_hard_dict):
    plt.plot(list(time_hard_dict[name]), c=colors[i])
    plt.legend(list(namedict.keys()))
plt.savefig('../fig/hard-all-time.pdf')

plt.clf()
plt.xlabel('iterations', fontsize=20)
plt.ylabel('execution time (milliseconds)', fontsize=20)
plt.title('without trap', fontsize=20)
for i, name in enumerate(time_hard_notrap_dict):
    plt.plot(list(time_hard_notrap_dict[name]), c=colors[i])
    plt.legend(list(namedict.keys()))
plt.savefig('../fig/hard-notrap-all-time.pdf')
```
